```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.preprocessing import normalize
import nltk
from textblob import TextBlob
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

## Clustered Comments (10 clusters, chosen arbritrarily)

```python
df = pd.read_csv('tiktok_comments.csv')

vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
X = vectorizer.fit_transform(df['text'])

n_clusters = 10
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(X)

df['cluster'] = kmeans.labels_


top_comments_per_cluster = 3

for cluster in range(n_clusters):
    cluster_data = df[df['cluster'] == cluster]
    num_comments = len(cluster_data)
    print(f"\nCluster #{cluster + 1} ({num_comments} comments):")
    top_comments = cluster_data.nlargest(top_comments_per_cluster, 'diggCount')
    for index, row in top_comments.iterrows():
        print(f"- {row['text']} (Likes: {row['diggCount']})")
```

## Comments Grouped by Sentiment, lowest to highest

```python
df['sentiment'] = df['text'].apply(lambda comment: TextBlob(str(comment)).sentiment.polarity)

df_sorted = df.sort_values(by='sentiment')

for index, row in df_sorted.iterrows():
    sentiment = "neutral"
    if row['sentiment'] > 0:
        sentiment = "positive"
    elif row['sentiment'] < 0:
        sentiment = "negative"
    print(f"{row['text']} (Sentiment: {sentiment} - Score: {row['sentiment']:.2f})\n")
```

```python
from fuzzywuzzy import fuzz


def contains_keywords(comment):
    keywords = ['friends', 'sharing']
    for keyword in keywords:
        if fuzz.partial_ratio(comment.lower(), keyword) > 65:
            return True
    return False
```

## Comments related to sharing and friends

```python
filtered_comments = df[df['text'].apply(contains_keywords)]

for index, row in filtered_comments.iterrows():
    print(f"- {row['text']} (Likes: {row['diggCount']})")
```
